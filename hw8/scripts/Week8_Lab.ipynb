for loop with list compression

```python
for fileid in gb.fileids():
    num_chars = len(gb.raw(fileid))
    num_words = len(gb.words(fileid))
    num_sents = len(gb.sents(fileid))
    num_vocab = len(set(w.lower() for w in gb.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

Data stored as a Data Frame

```python
import pandas as pd
np_object=[]
for fileid in gb.fileids():
    num_chars = len(gb.raw(fileid))
    num_words = len(gb.words(fileid))
    num_sents = len(gb.sents(fileid))
    num_vocab = len(set(w.lower() for w in gb.words(fileid)))
    x= [fileid,
        round(num_chars/num_words), 
        round(num_words/num_sents), 
        round(num_words/num_vocab)
       ]
    np_object.append(x)

df= pd.DataFrame(np_object, columns=['Title', '#chars/#words', '#words/#sents', '#words/#vocab'])
df
```

```python
import nltk
from nltk.corpus import gutenberg as gb

print ("")
print ("Relative frequencies of modals in 18 Gutenberg Corpus")
print ("")
titles = gb.fileids()
modals = ['can', 'could', 'may', 'might', 'will', 'should']
cfd = nltk.ConditionalFreqDist(
    (text, word)
    for text in gb.fileids()
    for word in gb.words(text))
cfd.tabulate(conditions=titles, samples=modals)
cfd
```

```python
dianes_grammar = nltk.CFG.fromstring("""
S -> VB
VB -> "Run"
""")
dianes_grammar
```

```python
print ("-------------------------")
print ("dianes_grammar")
print ("-------------------------")

text = nltk.word_tokenize("Run")

print ("Tree Output ")
print ("-------------------------")
rd_parser = nltk.RecursiveDescentParser(dianes_grammar, trace=0)
for tree in rd_parser.parse(text):
    print(tree)

# Try to draw a tree
# tree.draw()

exit(0)
```
